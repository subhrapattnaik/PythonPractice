```python
#compute the x and y coordinates for points on a sine curve

x = np.arange (0 , 4 * np.pi ,0.1) 
# 0.1 is the step size

y = np.sin(x)
plt.title("Sine wave form")

plt.plot(x,y)
```

```python
#subplot
#compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5* np.pi ,0.1)
y_sin =np.sin(x)
y_cos =np.cos(x)

#Setup a subplot grid that has height 2 and width 1
#and setup the first such subplot as active
plt.subplot(2,1,1)

#make the first plot
plt.plot(x,y_sin)
plt.title("Sine")

#set the second subplot as active , and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title("Cosine")

plt.show()
          
```

```python
#Bar Plot

x = [2,8,10]
y= [11,16,9]



plt.bar(x,y,align="center",color='g')
#by default color is blue
plt.ylabel('Y axis ')
plt.xlabel('X axis ')
plt.title('Bar graph ')

plt.show()
#%matplotlib inline # no need to write plt.show if you write in jupyter notebook
```

```python
#Box Plot
#helps to find percentile
# std is std deviation
# o o o circles are outliers
#0th (boottom line),25 percentile, 50th percentile,75 percentile,100th percentile
data = [np.random.normal(0,std,100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
#try with vert=False - it will be horizontal
#try with patch_artist=False -it will be without colour
```

```python
#Pie Chart
#Data to plot
labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','green','lightcoral','lightskyblue']
explode=(0,0.2,0,0) #explode first slice

#explode is for how much the slice will be moved
# 0.1 can be any number and which slice you want to move(if its on 1st position here - first slice)

#plot 
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

#autopct='%1.1f%% for float
#shadow=True for creating shadows
plt.axis('equal')
plt.show()
```
